```python
# customer segmentation
sns.histplot(shopping_trends['Age'], bins=10, kde=True)# Age distribution using histogram
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()


sns.countplot(data=shopping_trends, x='Gender')# Gender Distribution using barplot
plt.title("Gender Distribution")
plt.show()


location_freq = shopping_trends['Location'].value_counts()
location_freq.plot(kind='bar', figsize=(10, 5), title="Location-wise Purchase Frequency")# Location-based purchase frequency
plt.xlabel("Location")
plt.ylabel("Number of Purchases")
plt.show()

sns.histplot(shopping_trends['Frequency of Purchases'], kde=True)# Frequency of Purchases Distribution using histogram
plt.title('Frequency of Purchases Distribution')
plt.show()
```

```python
#purchase behaviour analysis
sns.boxplot(x='Gender', y='Purchase Amount (USD)', data=shopping_trends)# Purchase Amount Distribution analysis using boxplot
plt.title("Purchase Amount by Gender")
plt.show()


sns.countplot(data=shopping_trends, x='Preferred Payment Method')#analysis using countplot
plt.title("Preferred Payment Method")
plt.xticks(rotation=45)
plt.show()


sns.boxplot(x='Payment Method', y='Purchase Amount (USD)', data=shopping_trends)# Payment Method vs Purchase Amount analysis using boxplot
plt.title("Purchase Amount by Payment Method")
plt.xticks(rotation=45)
plt.show()
```

```python

sns.barplot(data=shopping_trends, x='Category', y='Purchase Amount (USD)', estimator=sum)# Sales by product category analysis using barplot
plt.title("Total Sales by Category")
plt.xticks(rotation=45)
plt.show()


sns.boxplot(data=shopping_trends, x='Category', y='Review Rating')# Review Ratings analysis using boxplot
plt.title("Review Ratings Distribution per Category")
plt.xticks(rotation=45)
plt.show()


sns.boxplot(x='Size', y='Purchase Amount (USD)', data=shopping_trends)
plt.title("Purchase Amount by Size")
plt.show()

sns.boxplot(x='Color', y='Purchase Amount (USD)', data=shopping_trends)
plt.title("Purchase Amount by Color")
plt.xticks(rotation=90)
plt.show()
```

```python

sns.scatterplot(x='Discount Applied', y='Purchase Amount (USD)', data=shopping_trends)#analysis using scatterplot
plt.title("Discount vs Purchase Amount")
plt.show()


sns.boxplot(data=shopping_trends, x='Promo Code Used', y='Purchase Amount (USD)')# promo code asage analysis using boxplot
plt.title("Effect of Promo Code Use on Purchase Amount")
plt.show()
```

```python

sns.countplot(x='Previous Purchases', hue='Subscription Status', data=shopping_trends)# previous purchase vs subscription analysis using countplot
plt.title("Previous Purchases by Subscription Status")
plt.show()


sns.boxplot(x='Subscription Status', y='Frequency of Purchases', data=shopping_trends)# subscription vs frequency of purchase using boxplot
plt.title("Frequency of Purchases by Subscription")
plt.show()
```

```python

sns.histplot(shopping_trends['Review Rating'], kde=True, bins=10)# review rating distribution analysis using histogram
plt.title("Review Rating Distribution")
plt.show()


sns.boxplot(data=shopping_trends, x='Shipping Type', y='Review Rating')# review ratings by shipping Type analysis using boxplot
plt.title("Review Ratings by Shipping Type")
plt.xticks(rotation=45)
plt.show()
```

```python

plt.figure(figsize=(12, 6))
sns.countplot(data=shopping_trends, y='Shipping Type', hue='Location')# shipping type preferences by location using countplot
plt.title("Shipping Type Preference by Location")
plt.legend(loc='lower right')
plt.show()


plt.figure(figsize=(12, 6))
sns.countplot(data=shopping_trends, x='Payment Method', hue='Location')# payment method popularity by Location using countplot
plt.title("Payment Methods by Location")
plt.xticks(rotation=45)
plt.show()
```

```python

sns.barplot(data=shopping_trends, x='Season', y='Purchase Amount (USD)', estimator=sum)#sales across sessons using barplot
plt.title('Total Sales across Seasons')
plt.show()
```

```python

numeric_shopping_trends = shopping_trends.copy()# convert categorical to numerical for correlation
numeric_shopping_trends['Promo Code Used'] = numeric_shopping_trends['Promo Code Used'].astype('category').cat.codes
numeric_shopping_trends['Subscription Status'] = numeric_shopping_trends['Subscription Status'].astype('category').cat.codes
numeric_shopping_trends['Preferred Payment Method'] = numeric_shopping_trends['Preferred Payment Method'].astype('category').cat.codes


num_cols = numeric_shopping_trends.select_dtypes(include=['int64', 'float64']).corr()# select only numeric columns


plt.figure(figsize=(10, 6))
sns.heatmap(num_cols, annot=True, cmap='viridis')# correlation heatmap
plt.title("Correlation Heatmap")
plt.show()
```
